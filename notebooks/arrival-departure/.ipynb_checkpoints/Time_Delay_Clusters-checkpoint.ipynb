# Clusters based on Delay time

## Full-data profile â€” columns, total rows

```python
import duckdb as ddb
import pyarrow.parquet as pq
import pandas as pd
import numpy as np
import os

PARQUET_PATH = globals().get(
    "PARQUET_PATH",
    "/content/drive/MyDrive/MBTA_Bus_Arrival_Departure_Times/MBTA Bus Arrival Departure Times/merged_output/mbta_arrival_departure_merged_direction.parquet"
)
assert os.path.exists(PARQUET_PATH), f"File not found: {PARQUET_PATH}"

pf = pq.ParquetFile(PARQUET_PATH)
arrow_schema = pf.schema.to_arrow_schema()
cols = [f.name for f in arrow_schema]
count_exprs = [ 'COUNT(*) AS total_rows' ]
for c in cols:
    count_exprs.append(f'COUNT("{c}") AS nn_{c}')

sql = f"""
SELECT
  {', '.join(count_exprs)}
FROM read_parquet('{PARQUET_PATH}')
"""

con = ddb.connect()
row = con.execute(sql).fetchdf().iloc[0].to_dict()

total_rows = int(row['total_rows'])
summary_rows = []
for c in cols:
    non_null = int(row[f'nn_{c}'])
    nulls = total_rows - non_null
    null_pct = (nulls / total_rows * 100.0) if total_rows > 0 else np.nan
    dtype = str(arrow_schema.field(c).type) if hasattr(arrow_schema, "field") else str(arrow_schema.types[cols.index(c)])
    summary_rows.append({
        "column": c,
        "dtype": dtype,
        "non_null": non_null,
        "nulls": nulls,
        "null_pct": round(null_pct, 3),
    })

profile_df = pd.DataFrame(summary_rows).sort_values("column").reset_index(drop=True)

print(f"Parquet path: {PARQUET_PATH}")
print(f"Total rows: {total_rows:,}")
print(f"Total columns: {len(cols)}")
print("\nColumns & dtypes:")
print(", ".join(f"{c} ({str(arrow_schema.field(c).type)})" for c in cols))

print("\nMissing values summary (sorted by highest null %):")
display(profile_df.sort_values("null_pct", ascending=False))
```

## Build views that handle NULLs the right way

```python
import duckdb as ddb
import os
PARQUET_PATH = globals().get(
    "PARQUET_PATH",
    "/content/drive/MyDrive/MBTA_Bus_Arrival_Departure_Times/MBTA Bus Arrival Departure Times/merged_output/mbta_arrival_departure_merged_direction.parquet"
)
OUT_DIR = globals().get("OUT_DIR", "/content/drive/MyDrive/mbta/outputs")
os.makedirs(OUT_DIR, exist_ok=True)

con = ddb.connect()

sql_make_views = f"""
-- 1) Base parse: robust timestamp casts from string
CREATE OR REPLACE VIEW mbta_base AS
SELECT
  route_id,
  lower(trim(coalesce(point_type,''))) AS pt,
  stop_id,
  lower(trim(coalesce(direction,''))) AS direction,
  service_date,
  -- normalize ISO strings before cast
  CAST(replace(replace(trim(scheduled),'T',' '),'Z','') AS TIMESTAMP) AS scheduled_ts,
  CAST(replace(replace(trim(actual),'T',' '),'Z','')    AS TIMESTAMP) AS actual_ts,
  scheduled_headway,  -- may be null
  headway             -- may be null
FROM read_parquet('{PARQUET_PATH}');

-- 2) Lateness-ready rows: require both timestamps; clip extreme delays
CREATE OR REPLACE VIEW mbta_clean_late AS
WITH delays AS (
  SELECT
    route_id, pt, stop_id, direction, service_date,
    DATE_DIFF('second', scheduled_ts, actual_ts) / 60.0 AS delay_min
  FROM mbta_base
  WHERE scheduled_ts IS NOT NULL AND actual_ts IS NOT NULL
)
SELECT *
FROM delays
WHERE delay_min BETWEEN -30 AND 120;

-- 3) Headway reconstruction (best-available): per route+direction+stop+date
--    We keep provided headways when present; otherwise compute with LAG().
CREATE OR REPLACE VIEW mbta_headway_best AS
WITH ordered AS (
  SELECT
    route_id, direction, stop_id, service_date,
    scheduled_ts, actual_ts,
    scheduled_headway, headway,
    LAG(actual_ts)    OVER (PARTITION BY route_id, direction, stop_id, service_date ORDER BY actual_ts)    AS prev_actual_ts,
    LAG(scheduled_ts) OVER (PARTITION BY route_id, direction, stop_id, service_date ORDER BY scheduled_ts) AS prev_scheduled_ts
  FROM mbta_base
  WHERE (actual_ts IS NOT NULL OR scheduled_ts IS NOT NULL)
)
SELECT
  route_id, direction, stop_id, service_date,
  -- observed headway in minutes (prefer provided, else reconstruct)
  COALESCE(headway,
           CASE WHEN prev_actual_ts IS NOT NULL AND actual_ts IS NOT NULL
                THEN DATE_DIFF('second', prev_actual_ts, actual_ts)/60.0
                ELSE NULL END)                                       AS headway_obs_best,
  -- scheduled headway in minutes (prefer provided, else reconstruct)
  COALESCE(scheduled_headway,
           CASE WHEN prev_scheduled_ts IS NOT NULL AND scheduled_ts IS NOT NULL
                THEN DATE_DIFF('second', prev_scheduled_ts, scheduled_ts)/60.0
                ELSE NULL END)                                       AS headway_sched_best
FROM ordered
-- keep rows where we have at least one usable headway
WHERE headway IS NOT NULL
   OR (prev_actual_ts IS NOT NULL AND actual_ts IS NOT NULL)
   OR (prev_scheduled_ts IS NOT NULL AND scheduled_ts IS NOT NULL);
"""

con.execute(sql_make_views)
print("Views created: mbta_base, mbta_clean_late, mbta_headway_best âœ…")
n_late = con.execute("SELECT COUNT(*) FROM mbta_clean_late").fetchone()[0]
print("Rows usable for delay:", f"{n_late:,}")
```

```python
import os, duckdb as ddb

OUT_DIR = globals().get("OUT_DIR", "/content/drive/MyDrive/mbta/outputs")
os.makedirs(OUT_DIR, exist_ok=True)
DUCKDB_PATH = os.path.join(OUT_DIR, "mbta.duckdb")

con = ddb.connect(DUCKDB_PATH)
print("Connected to:", DUCKDB_PATH)
PARQUET_PATH = globals().get(
    "PARQUET_PATH",
    "/content/drive/MyDrive/MBTA_Bus_Arrival_Departure_Times/MBTA Bus Arrival Departure Times/merged_output/mbta_arrival_departure_merged_direction.parquet"
)

sql_make_views = f"""
CREATE OR REPLACE VIEW mbta_base AS
SELECT
  route_id,
  lower(trim(coalesce(point_type,''))) AS pt,
  stop_id,
  lower(trim(coalesce(direction,''))) AS direction,
  service_date,
  CAST(replace(replace(trim(scheduled),'T',' '),'Z','') AS TIMESTAMP) AS scheduled_ts,
  CAST(replace(replace(trim(actual),'T',' '),'Z','')    AS TIMESTAMP) AS actual_ts,
  scheduled_headway,
  headway
FROM read_parquet('{PARQUET_PATH}');

CREATE OR REPLACE VIEW mbta_clean_late AS
WITH delays AS (
  SELECT
    route_id, pt, stop_id, direction, service_date,
    DATE_DIFF('second', scheduled_ts, actual_ts) / 60.0 AS delay_min
  FROM mbta_base
  WHERE scheduled_ts IS NOT NULL AND actual_ts IS NOT NULL
)
SELECT *
FROM delays
WHERE delay_min BETWEEN -30 AND 120;

CREATE OR REPLACE VIEW mbta_headway_best AS
WITH ordered AS (
  SELECT
    route_id, direction, stop_id, service_date,
    scheduled_ts, actual_ts,
    scheduled_headway, headway,
    LAG(actual_ts)    OVER (PARTITION BY route_id, direction, stop_id, service_date ORDER BY actual_ts)    AS prev_actual_ts,
    LAG(scheduled_ts) OVER (PARTITION BY route_id, direction, stop_id, service_date ORDER BY scheduled_ts) AS prev_scheduled_ts
  FROM mbta_base
  WHERE (actual_ts IS NOT NULL OR scheduled_ts IS NOT NULL)
)
SELECT
  route_id, direction, stop_id, service_date,
  COALESCE(headway,
           CASE WHEN prev_actual_ts IS NOT NULL AND actual_ts IS NOT NULL
                THEN DATE_DIFF('second', prev_actual_ts, actual_ts)/60.0
                ELSE NULL END) AS headway_obs_best,
  COALESCE(scheduled_headway,
           CASE WHEN prev_scheduled_ts IS NOT NULL AND scheduled_ts IS NOT NULL
                THEN DATE_DIFF('second', prev_scheduled_ts, scheduled_ts)/60.0
                ELSE NULL END) AS headway_sched_best
FROM ordered
WHERE headway IS NOT NULL
   OR (prev_actual_ts IS NOT NULL AND actual_ts IS NOT NULL)
   OR (prev_scheduled_ts IS NOT NULL AND scheduled_ts IS NOT NULL);
"""
con.execute(sql_make_views)

n_late = con.execute("SELECT COUNT(*) FROM mbta_clean_late").fetchone()[0]
print("Views ready. Rows usable for delay:", f"{n_late:,}")
```

```python
import os, pandas as pd, matplotlib.pyplot as plt

OUT_DIR = globals().get("OUT_DIR", "/content/drive/MyDrive/mbta/outputs")
os.makedirs(OUT_DIR, exist_ok=True)

LATE_THRESHOLD_MIN = 5.0
MIN_OBS_PER_ROUTE  = 300

sql_late = f"""
WITH end_agg AS (
  SELECT
    route_id,
    COUNT(*) AS num_obs_end,
    AVG(CASE WHEN delay_min > {LATE_THRESHOLD_MIN} THEN 1 ELSE 0 END)::DOUBLE AS late_rate_end,
    AVG(delay_min) AS avg_delay_end
  FROM mbta_clean_late
  WHERE pt = 'endpoint'
  GROUP BY route_id
),
all_agg AS (
  SELECT
    route_id,
    COUNT(*) AS num_obs_all,
    AVG(CASE WHEN delay_min > {LATE_THRESHOLD_MIN} THEN 1 ELSE 0 END)::DOUBLE AS late_rate_all,
    AVG(delay_min) AS avg_delay_all
  FROM mbta_clean_late
  GROUP BY route_id
)
SELECT
  COALESCE(e.route_id, a.route_id) AS route_id,
  e.num_obs_end, e.late_rate_end, e.avg_delay_end,
  a.num_obs_all, a.late_rate_all, a.avg_delay_all,
  CASE WHEN COALESCE(e.num_obs_end,0) >= {MIN_OBS_PER_ROUTE} THEN e.late_rate_end ELSE a.late_rate_all END AS late_rate_best,
  CASE WHEN COALESCE(e.num_obs_end,0) >= {MIN_OBS_PER_ROUTE} THEN e.avg_delay_end  ELSE a.avg_delay_all  END AS avg_delay_best,
  CASE WHEN COALESCE(e.num_obs_end,0) >= {MIN_OBS_PER_ROUTE} THEN 'endpoint' ELSE 'all' END AS late_rate_source
FROM end_agg e
FULL OUTER JOIN all_agg a USING (route_id)
WHERE COALESCE(e.num_obs_end,0) >= {MIN_OBS_PER_ROUTE}
   OR COALESCE(a.num_obs_all,0) >= {MIN_OBS_PER_ROUTE}
ORDER BY late_rate_best DESC;
"""
late_routes = con.execute(sql_late).fetchdf()
print("Routes with sufficient observations:", late_routes.shape[0])
display(late_routes.head(10))

out_csv = os.path.join(OUT_DIR, "late_rate_per_route.csv")
late_routes.to_csv(out_csv, index=False)
print("Saved:", out_csv)
plt.rcParams.update({
    "figure.figsize": (10, 6), "figure.dpi": 120,
    "axes.grid": True, "grid.alpha": 0.25,
    "axes.spines.top": False, "axes.spines.right": False,
    "axes.titlesize": 16, "axes.labelsize": 12,
    "xtick.labelsize": 11, "ytick.labelsize": 11,
    "legend.frameon": False, "axes.axisbelow": True, "font.size": 12,
})
top = late_routes.sort_values("late_rate_best", ascending=False).head(15).copy()
top["route_id"] = top["route_id"].astype(str)
top_plot = top.sort_values("late_rate_best", ascending=True)
vals = (top_plot["late_rate_best"] * 100.0).values
labels = top_plot["route_id"].values
plt.barh(labels, vals)
for y, v in enumerate(vals):
    plt.text(v + 0.6, y, f"{v:.1f}%", va="center")
plt.title("Which lines are late more often?  (Top 15 by late rate > 5 min)")
plt.xlabel("Late rate (% of records)"); plt.ylabel("Route ID")
plt.tight_layout(); plt.show()

display(
  top_plot[["route_id","late_rate_best","avg_delay_best","late_rate_source","num_obs_end","num_obs_all"]]
    .sort_values("late_rate_best", ascending=False)
)
```

## Lateness per route Ã— year (Endpoint-first with All fallback)

```python
import os, duckdb as ddb, pandas as pd, numpy as np, matplotlib.pyplot as plt

OUT_DIR = globals().get("OUT_DIR", "/content/drive/MyDrive/mbta/outputs")
os.makedirs(OUT_DIR, exist_ok=True)
DUCKDB_PATH = os.path.join(OUT_DIR, "mbta.duckdb")

con = ddb.connect(DUCKDB_PATH)
print("Using DB:", DUCKDB_PATH)

PARQUET_PATH = globals().get(
    "PARQUET_PATH",
    "/content/drive/MyDrive/MBTA_Bus_Arrival_Departure_Times/MBTA Bus Arrival Departure Times/merged_output/mbta_arrival_departure_merged_direction.parquet"
)
con.execute(f"""
CREATE OR REPLACE VIEW mbta_base AS
SELECT
  route_id,
  lower(trim(coalesce(point_type,''))) AS pt,
  stop_id,
  lower(trim(coalesce(direction,''))) AS direction,
  service_date,
  CAST(replace(replace(trim(scheduled),'T',' '),'Z','') AS TIMESTAMP) AS scheduled_ts,
  CAST(replace(replace(trim(actual),'T',' '),'Z','')    AS TIMESTAMP) AS actual_ts,
  scheduled_headway,
  headway
FROM read_parquet('{PARQUET_PATH}');

CREATE OR REPLACE VIEW mbta_clean_late AS
WITH delays AS (
  SELECT
    route_id, pt, stop_id, direction, service_date,
    DATE_DIFF('second', scheduled_ts, actual_ts) / 60.0 AS delay_min
  FROM mbta_base
  WHERE scheduled_ts IS NOT NULL AND actual_ts IS NOT NULL
)
SELECT *
FROM delays
WHERE delay_min BETWEEN -30 AND 120;
""")

LATE_THRESHOLD_MIN = 5.0
MIN_OBS_PER_ROUTE_YEAR = 200

sql_year = f"""
WITH base AS (
  SELECT
    route_id, pt,
    TRY_CAST(service_date AS DATE) AS service_d,
    delay_min
  FROM mbta_clean_late
  WHERE service_date IS NOT NULL
),
end_year AS (
  SELECT
    route_id,
    EXTRACT(YEAR FROM service_d)::INT AS year,
    COUNT(*) AS n_obs_end,
    AVG(CASE WHEN delay_min > {LATE_THRESHOLD_MIN} THEN 1 ELSE 0 END)::DOUBLE AS late_rate_end,
    AVG(delay_min) AS avg_delay_end
  FROM base
  WHERE pt = 'endpoint' AND service_d IS NOT NULL
  GROUP BY route_id, year
),
all_year AS (
  SELECT
    route_id,
    EXTRACT(YEAR FROM service_d)::INT AS year,
    COUNT(*) AS n_obs_all,
    AVG(CASE WHEN delay_min > {LATE_THRESHOLD_MIN} THEN 1 ELSE 0 END)::DOUBLE AS late_rate_all,
    AVG(delay_min) AS avg_delay_all
  FROM base
  WHERE service_d IS NOT NULL
  GROUP BY route_id, year
),
combined AS (
  SELECT
    COALESCE(e.route_id, a.route_id) AS route_id,
    COALESCE(e.year, a.year)         AS year,
    e.n_obs_end, e.late_rate_end, e.avg_delay_end,
    a.n_obs_all, a.late_rate_all, a.avg_delay_all,
    CASE WHEN COALESCE(e.n_obs_end,0) >= {MIN_OBS_PER_ROUTE_YEAR} THEN e.late_rate_end ELSE a.late_rate_all END AS late_rate_best,
    CASE WHEN COALESCE(e.n_obs_end,0) >= {MIN_OBS_PER_ROUTE_YEAR} THEN e.avg_delay_end  ELSE a.avg_delay_all  END AS avg_delay_best,
    CASE WHEN COALESCE(e.n_obs_end,0) >= {MIN_OBS_PER_ROUTE_YEAR} THEN e.n_obs_end      ELSE a.n_obs_all      END AS n_obs_best,
    CASE WHEN COALESCE(e.n_obs_end,0) >= {MIN_OBS_PER_ROUTE_YEAR} THEN 'endpoint'       ELSE 'all'             END AS source
  FROM end_year e
  FULL OUTER JOIN all_year a USING (route_id, year)
)
SELECT *
FROM combined
WHERE COALESCE(n_obs_end,0) >= {MIN_OBS_PER_ROUTE_YEAR}
   OR COALESCE(n_obs_all,0) >= {MIN_OBS_PER_ROUTE_YEAR}
ORDER BY year, late_rate_best DESC;
"""

route_year = con.execute(sql_year).fetchdf()
print("route_year shape:", route_year.shape)
display(route_year.head(12))

out_csv = os.path.join(OUT_DIR, "route_lateness_by_year_best.csv")
route_year.to_csv(out_csv, index=False)
print("Saved:", out_csv)

plt.rcParams.update({
    "figure.figsize": (10, 6), "figure.dpi": 120,
    "axes.grid": True, "grid.alpha": 0.25,
    "axes.spines.top": False, "axes.spines.right": False,
    "axes.titlesize": 16, "axes.labelsize": 12,
    "xtick.labelsize": 11, "ytick.labelsize": 11,
    "legend.frameon": False, "axes.axisbelow": True, "font.size": 12,
})

if not route_year.empty:
    latest_year = int(route_year["year"].max())
    ry_latest = route_year[route_year["year"] == latest_year].copy()
    top = ry_latest.sort_values("late_rate_best", ascending=False).head(15).copy()
    top["route_id"] = top["route_id"].astype(str)
    top_sort = top.sort_values("late_rate_best", ascending=True)
    vals = (top_sort["late_rate_best"] * 100.0).values
    labels = top_sort["route_id"].values
    plt.barh(labels, vals)
    for y, v in enumerate(vals):
        plt.text(v + 0.6, y, f"{v:.1f}%", va="center")
    plt.title(f"Most-late routes in {latest_year}  (late > {LATE_THRESHOLD_MIN:.0f} min)")
    plt.xlabel("Late rate (% of records)"); plt.ylabel("Route ID")
    plt.tight_layout(); plt.show()

    top5_routes = top.sort_values("late_rate_best", ascending=False)["route_id"].head(5).tolist()
    trend = route_year[route_year["route_id"].astype(str).isin(top5_routes)].copy()
    years = sorted(route_year["year"].unique())
    plt.figure(figsize=(11,6))
    for rid in top5_routes:
        s = trend[trend["route_id"].astype(str) == rid].set_index("year").reindex(years)
        plt.plot(years, s["late_rate_best"] * 100.0, marker="o", label=str(rid))
    plt.title("Late-rate trends by year (Top-5 worst routes)")
    plt.xlabel("Year"); plt.ylabel("Late rate (%)")
    plt.legend(title="Route"); plt.tight_layout(); plt.show()
    overall = (route_year.assign(wlate=route_year["late_rate_best"] * route_year["n_obs_best"])
                         .groupby("route_id")[["wlate","n_obs_best"]].sum())
    overall["late_rate_weighted"] = overall["wlate"] / overall["n_obs_best"]
    worst20 = overall.sort_values("late_rate_weighted", ascending=False).head(20).index

    mat = (route_year[route_year["route_id"].isin(worst20)]
              .pivot_table(index="route_id", columns="year", values="late_rate_best")).loc[worst20]
    plt.figure(figsize=(12, 7))
    im = plt.imshow(mat.values * 100.0, aspect="auto")
    plt.colorbar(im, label="Late rate (%)")
    plt.yticks(ticks=np.arange(mat.shape[0]), labels=mat.index.astype(str))
    plt.xticks(ticks=np.arange(mat.shape[1]), labels=mat.columns.astype(int))
    plt.title("Late rate (%) by route Ã— year â€” worst 20 routes overall")
    plt.xlabel("Year"); plt.ylabel("Route ID")
    plt.tight_layout(); plt.show()
else:
    print("No route-year groups met the minimum observation threshold. Consider lowering MIN_OBS_PER_ROUTE_YEAR.")
```

## Clusters on overall lateness : KMeans (random), KMeans++ and DBSCAN, pick best

```python
# === Cell D-viz: Visuals for overall clusters (k-means, k-means++, DBSCAN) ===
import os, numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

OUT_DIR = globals().get("OUT_DIR", "/content/drive/MyDrive/mbta/outputs")
FIG_DIR = os.path.join(OUT_DIR, "clusters_overall", "figures")
os.makedirs(FIG_DIR, exist_ok=True)

# Reuse X, X_scaled, ids, results, res_df from Cell D; if missing, rebuild quickly
need_rebuild = False
for name in ["X", "X_scaled", "ids"]:
    if name not in globals():
        need_rebuild = True
        break

if need_rebuild or "results" not in globals():
    print("Recomputing features and candidates (lightweight)...")
    # Pull the per-route table again (same SQL as Cell D but faster than reading parquet)
    import duckdb as ddb
    DUCKDB_PATH = os.path.join(OUT_DIR, "mbta.duckdb")
    con = ddb.connect(DUCKDB_PATH)
    LATE_THRESHOLD_MIN = 5.0
    MIN_OBS_PER_ROUTE  = 300

    sql_late = f"""
    WITH end_agg AS (
      SELECT route_id,
             COUNT(*) AS num_obs_end,
             AVG(CASE WHEN delay_min > {LATE_THRESHOLD_MIN} THEN 1 ELSE 0 END)::DOUBLE AS late_rate_end,
             AVG(delay_min) AS avg_delay_end
      FROM mbta_clean_late
      WHERE pt = 'endpoint'
      GROUP BY route_id
    ),
    all_agg AS (
      SELECT route_id,
             COUNT(*) AS num_obs_all,
             AVG(CASE WHEN delay_min > {LATE_THRESHOLD_MIN} THEN 1 ELSE 0 END)::DOUBLE AS late_rate_all,
             AVG(delay_min) AS avg_delay_all
      FROM mbta_clean_late
      GROUP BY route_id
    )
    SELECT
      COALESCE(e.route_id, a.route_id) AS route_id,
      CASE WHEN COALESCE(e.num_obs_end,0) >= {MIN_OBS_PER_ROUTE} THEN e.late_rate_end ELSE a.late_rate_all END AS late_rate_best,
      CASE WHEN COALESCE(e.num_obs_end,0) >= {MIN_OBS_PER_ROUTE} THEN e.avg_delay_end  ELSE a.avg_delay_all  END AS avg_delay_best
    FROM end_agg e
    FULL OUTER JOIN all_agg a USING (route_id)
    WHERE COALESCE(e.num_obs_end,0) >= {MIN_OBS_PER_ROUTE}
       OR COALESCE(a.num_obs_all,0) >= {MIN_OBS_PER_ROUTE}
    """
    late_routes = con.execute(sql_late).fetchdf()

    feat_cols = ["late_rate_best", "avg_delay_best"]
    X = late_routes[feat_cols].apply(pd.to_numeric, errors="coerce").dropna().copy()
    ids = late_routes.loc[X.index, "route_id"].astype(str).values
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    def eval_labels(Xs, labels):
        uniq = np.unique(labels)
        if len(uniq) <= 1:
            return {"sil": -1.0, "dbi": np.inf, "chi": 0.0}
        try:
            return {
                "sil": silhouette_score(Xs, labels),
                "dbi": davies_bouldin_score(Xs, labels),
                "chi": calinski_harabasz_score(Xs, labels),
            }
        except Exception:
            return {"sil": -1.0, "dbi": np.inf, "chi": 0.0}

    results = []
    # KMeans random & ++ sweeps
    for init_mode in ["random", "k-means++"]:
        for k in range(2, 7):
            km = KMeans(n_clusters=k, init=init_mode, n_init=10, random_state=42)
            lab = km.fit_predict(X_scaled)
            m = eval_labels(X_scaled, lab)
            results.append({"method": "kmeans_random" if init_mode=="random" else "kmeans_plus", "k": k, **m, "labels": lab})
    # DBSCAN grid
    for eps in [0.3, 0.5, 0.7, 1.0, 1.3]:
        for ms in [3,4,5]:
            db = DBSCAN(eps=eps, min_samples=ms)
            lab = db.fit_predict(X_scaled)
            m = eval_labels(X_scaled, lab)
            results.append({"method": "dbscan", "eps": eps, "min_samples": ms, **m, "labels": lab})

res_df = pd.DataFrame([{k:v for k,v in r.items() if k!="labels"} for r in results])

# ---- 1) Silhouette vs K for KMeans (random & ++)
plt.figure(figsize=(10,6))
for method, label in [("kmeans_random","k-means (random)"), ("kmeans_plus","k-means++")]:
    tmp = res_df[res_df["method"]==method].dropna(subset=["k","sil"]).sort_values("k")
    if not tmp.empty:
        plt.plot(tmp["k"], tmp["sil"], marker="o", label=label)
plt.title("Silhouette vs K â€” k-means variants")
plt.xlabel("K"); plt.ylabel("Silhouette")
plt.legend(); plt.grid(alpha=0.3)
plt.tight_layout(); plt.savefig(os.path.join(FIG_DIR, "silhouette_vs_k.png"), bbox_inches="tight"); plt.show()
```

```python
# ---- 2) DBSCAN grid heatmap (silhouette) ----
db = res_df[res_df["method"]=="dbscan"].dropna(subset=["eps","min_samples","sil"])
if not db.empty:
    piv = db.pivot_table(index="min_samples", columns="eps", values="sil", aggfunc="max")
    plt.figure(figsize=(8,4.8))
    im = plt.imshow(piv.values, aspect="auto")
    plt.colorbar(im, label="Silhouette")
    plt.yticks(range(len(piv.index)), piv.index)
    plt.xticks(range(len(piv.columns)), piv.columns)
    plt.xlabel("eps"); plt.ylabel("min_samples")
    plt.title("DBSCAN silhouette heatmap")
    plt.tight_layout(); plt.savefig(os.path.join(FIG_DIR, "dbscan_heatmap.png"), bbox_inches="tight"); plt.show()
```

```python
# ---- helper to render scatters ----
def scatter_for_labels(labels, title, fname):
    # order clusters by mean lateness (Low â†’ High)
    tmp = pd.DataFrame({
        "lr": X["late_rate_best"].values,
        "ad": X["avg_delay_best"].values,
        "c": np.asarray(labels)
    })
    order = (tmp.groupby("c")["lr"].mean()
               .sort_values()
               .index.tolist())
    rank = {cl:i for i,cl in enumerate(order)}
    plt.figure(figsize=(10,6))
    for cl in order:
        m = (tmp["c"].values == cl)
        name = "Low" if rank[cl]==0 else ("High" if rank[cl]==max(rank.values()) else "Mid")
        plt.scatter(tmp.loc[m,"lr"]*100.0, tmp.loc[m,"ad"], s=40, alpha=0.9, label=f"{name} (c{cl})")
    plt.xlabel("Late rate (>5 min) [%]"); plt.ylabel("Average delay [min]")
    plt.title(title); plt.legend()
    plt.tight_layout(); plt.savefig(os.path.join(FIG_DIR, fname), bbox_inches="tight"); plt.show()

# ---- 3) Best per method (k-means random, k-means++, DBSCAN) ----
best_per = {}
for method in ["kmeans_random","kmeans_plus","dbscan"]:
    cand = res_df[res_df["method"]==method]
    if cand.empty:
        continue
    idx = cand["sil"].idxmax()  # index in res_df aligns with `results`
    best_per[method] = {"row": cand.loc[idx].to_dict(), "labels": results[int(idx)]["labels"]}

for method, meta in best_per.items():
    r = meta["row"]
    t = (f"{'k-means (random)' if method=='kmeans_random' else 'k-means++' if method=='kmeans_plus' else 'DBSCAN'} "
         f"{'(k='+str(int(r['k']))+')' if 'k' in r and pd.notna(r.get('k')) else f'(eps={r.get('eps')}, min_samples={r.get('min_samples')})'} "
         f"â€” sil={r['sil']:.3f}, dbi={r['dbi']:.3f}, chi={r['chi']:.1f}")
    scatter_for_labels(meta["labels"], t, f"{method}_best_scatter.png")

    # cluster size bar
    sizes = pd.Series(np.asarray(meta["labels"])).value_counts().sort_index()
    plt.figure(figsize=(7,4))
    sizes.plot(kind="bar")
    plt.title(f"{method}: cluster sizes")
    plt.xlabel("Cluster id"); plt.ylabel("Routes")
    plt.tight_layout(); plt.savefig(os.path.join(FIG_DIR, f"{method}_cluster_sizes.png"), bbox_inches="tight"); plt.show()
```

```python
# ---- 4) Overall best (global max silhouette) ----
best_ix = int(res_df["sil"].idxmax())
best = results[best_ix]
desc = res_df.loc[best_ix].to_dict()
title = (f"OVERALL BEST â€” "
         f"{'k-means (random)' if desc['method']=='kmeans_random' else 'k-means++' if desc['method']=='kmeans_plus' else 'DBSCAN'} "
         f"{'(k='+str(int(desc['k']))+')' if 'k' in desc and pd.notna(desc.get('k')) else f'(eps={desc.get('eps')}, min_samples={desc.get('min_samples')})'} "
         f"â€” sil={desc['sil']:.3f}, dbi={desc['dbi']:.3f}, chi={desc['chi']:.1f}")
scatter_for_labels(best["labels"], title, "overall_best_scatter.png")

print("Figures saved to:", FIG_DIR)
```

## Clusters by year : run KMeans (random), KMeans++, DBSCAN per year; pick best each year

```python
import os, numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

OUT_DIR = globals().get("OUT_DIR", "/content/drive/MyDrive/mbta/outputs")
DUCKDB_PATH = os.path.join(OUT_DIR, "mbta.duckdb")
FIG_DIR = os.path.join(OUT_DIR, "clusters_by_year", "figures")
os.makedirs(FIG_DIR, exist_ok=True)

import duckdb as ddb
con = ddb.connect(DUCKDB_PATH)

LATE_THRESHOLD_MIN = 5.0
MIN_OBS_PER_ROUTE_YEAR = 200
MIN_ROUTES_TO_CLUSTER = 25

sql_year = f"""
WITH base AS (
  SELECT route_id, pt, TRY_CAST(service_date AS DATE) AS service_d, delay_min
  FROM mbta_clean_late
  WHERE service_date IS NOT NULL
),
end_year AS (
  SELECT route_id, EXTRACT(YEAR FROM service_d)::INT AS year,
         COUNT(*) AS n_obs_end,
         AVG(CASE WHEN delay_min > {LATE_THRESHOLD_MIN} THEN 1 ELSE 0 END)::DOUBLE AS late_rate_end,
         AVG(delay_min) AS avg_delay_end
  FROM base WHERE pt='endpoint' AND service_d IS NOT NULL
  GROUP BY route_id, year
),
all_year AS (
  SELECT route_id, EXTRACT(YEAR FROM service_d)::INT AS year,
         COUNT(*) AS n_obs_all,
         AVG(CASE WHEN delay_min > {LATE_THRESHOLD_MIN} THEN 1 ELSE 0 END)::DOUBLE AS late_rate_all,
         AVG(delay_min) AS avg_delay_all
  FROM base WHERE service_d IS NOT NULL
  GROUP BY route_id, year
),
combined AS (
  SELECT
    COALESCE(e.route_id,a.route_id) AS route_id,
    COALESCE(e.year,a.year) AS year,
    e.n_obs_end, e.late_rate_end, e.avg_delay_end,
    a.n_obs_all, a.late_rate_all, a.avg_delay_all,
    CASE WHEN COALESCE(e.n_obs_end,0) >= {MIN_OBS_PER_ROUTE_YEAR} THEN e.late_rate_end ELSE a.late_rate_all END AS late_rate_best,
    CASE WHEN COALESCE(e.n_obs_end,0) >= {MIN_OBS_PER_ROUTE_YEAR} THEN e.avg_delay_end  ELSE a.avg_delay_all  END AS avg_delay_best
  FROM end_year e FULL OUTER JOIN all_year a USING (route_id, year)
)
SELECT * FROM combined
WHERE COALESCE(n_obs_end,0) >= {MIN_OBS_PER_ROUTE_YEAR}
   OR COALESCE(n_obs_all,0) >= {MIN_OBS_PER_ROUTE_YEAR};
"""
route_year = con.execute(sql_year).fetchdf()
years = sorted(route_year["year"].dropna().unique())
print("Years:", years)

def eval_labels(Xs, labels):
    uniq = np.unique(labels)
    if len(uniq) <= 1:
        return {"sil": -1.0, "dbi": np.inf, "chi": 0.0}
    try:
        return {
            "sil": silhouette_score(Xs, labels),
            "dbi": davies_bouldin_score(Xs, labels),
            "chi": calinski_harabasz_score(Xs, labels),
        }
    except Exception:
        return {"sil": -1.0, "dbi": np.inf, "chi": 0.0}

score_rows = []
best_by_year = []
assign_by_year = {}

for yr in years:
    dfy = route_year[route_year["year"]==yr][["route_id","late_rate_best","avg_delay_best"]].dropna().copy()
    dfy["route_id"] = dfy["route_id"].astype(str)
    if dfy.shape[0] < MIN_ROUTES_TO_CLUSTER:
        print(f"Skip {yr}: only {dfy.shape[0]} routes")
        continue

    X = dfy[["late_rate_best","avg_delay_best"]].values
    Xs = StandardScaler().fit_transform(X)

    candidates = []
    for init_mode, tag in [("random","kmeans_random"), ("k-means++","kmeans_plus")]:
        for k in range(2,7):
            km = KMeans(n_clusters=k, init=init_mode, n_init=10, random_state=42)
            lab = km.fit_predict(Xs)
            m = eval_labels(Xs, lab)
            candidates.append({"year":yr, "method":tag, "k":k, **m, "labels":lab})
    for eps in [0.3,0.5,0.7,1.0,1.3]:
        for ms in [3,4,5]:
            db = DBSCAN(eps=eps, min_samples=ms)
            lab = db.fit_predict(Xs)
            m = eval_labels(Xs, lab)
            candidates.append({"year":yr, "method":"dbscan", "eps":eps, "min_samples":ms, **m, "labels":lab})

    scores = pd.DataFrame([{k:v for k,v in c.items() if k!="labels"} for c in candidates])
    score_rows.append(scores)

    best_idx = scores["sil"].idxmax()
    best = candidates[int(best_idx)]
    best_by_year.append({"year": yr, **{k:best.get(k) for k in ["method","k","eps","min_samples","sil","dbi","chi"]}})
    assign_by_year[yr] = {"labels": best["labels"], "dfy": dfy}

summary = pd.DataFrame(best_by_year)
if not summary.empty:
    plt.figure(figsize=(10,5))
    plt.bar(summary["year"].astype(str), summary["sil"])
    for i,(y,s) in enumerate(zip(summary["year"], summary["sil"])):
        plt.text(i, s+0.01, f"{s:.2f}", ha="center", va="bottom")
    plt.title("Best silhouette per year (winner method per year)")
    plt.xlabel("Year"); plt.ylabel("Silhouette")
    plt.tight_layout(); plt.savefig(os.path.join(FIG_DIR, "year_best_silhouette.png"), bbox_inches="tight"); plt.show()

    print("Winners by year:")
    display(summary[["year","method","k","eps","min_samples","sil","dbi","chi"]].sort_values("year"))

if len(assign_by_year)>0:
    latest = max(assign_by_year.keys())
    dfy = route_year[route_year["year"]==latest][["route_id","late_rate_best","avg_delay_best"]].dropna().copy()
    dfy["route_id"] = dfy["route_id"].astype(str)
    X = dfy[["late_rate_best","avg_delay_best"]].values
    Xs = StandardScaler().fit_transform(X)

    cand_latest = []
    for init_mode, tag in [("random","kmeans_random"), ("k-means++","kmeans_plus")]:
        for k in range(2,7):
            km = KMeans(n_clusters=k, init=init_mode, n_init=10, random_state=42)
            lab = km.fit_predict(Xs)
            m = eval_labels(Xs, lab)
            cand_latest.append({"method":tag, "k":k, **m, "labels":lab})
    for eps in [0.3,0.5,0.7,1.0,1.3]:
        for ms in [3,4,5]:
            db = DBSCAN(eps=eps, min_samples=ms)
            lab = db.fit_predict(Xs)
            m = eval_labels(Xs, lab)
            cand_latest.append({"method":"dbscan", "eps":eps, "min_samples":ms, **m, "labels":lab})
    s_latest = pd.DataFrame([{k:v for k,v in c.items() if k!="labels"} for c in cand_latest])

    def best_of(method):
        sub = s_latest[s_latest["method"]==method]
        if sub.empty: return None
        idx = sub["sil"].idxmax()
        cand = cand_latest[int(idx)]
        return cand

    best_r = best_of("kmeans_random")
    best_p = best_of("kmeans_plus")
    best_d = best_of("dbscan")

    def scatter_latest(labels, title, fname):
        if labels is None: return
        tmp = pd.DataFrame({"lr": dfy["late_rate_best"].values, "ad": dfy["avg_delay_best"].values, "c": labels})
        order = tmp.groupby("c")["lr"].mean().sort_values().index.tolist()
        rank = {cl:i for i,cl in enumerate(order)}
        plt.figure(figsize=(10,6))
        for cl in order:
            m = (labels == cl)
            name = "Low" if rank[cl]==0 else ("High" if rank[cl]==max(rank.values()) else "Mid")
            plt.scatter(tmp.loc[m,"lr"]*100.0, tmp.loc[m,"ad"], s=40, alpha=0.9, label=f"{name} (c{cl})")
        plt.xlabel("Late rate (>5 min) [%]"); plt.ylabel("Average delay [min]")
        plt.title(title); plt.legend()
        plt.tight_layout(); plt.savefig(os.path.join(FIG_DIR, fname), bbox_inches="tight"); plt.show()

    if best_r:
        scatter_latest(best_r["labels"], f"{latest}: k-means (random, k={best_r['k']}) sil={best_r['sil']:.3f}",
                       f"{latest}_kmeans_random_scatter.png")
    if best_p:
        scatter_latest(best_p["labels"], f"{latest}: k-means++ (k={best_p['k']}) sil={best_p['sil']:.3f}",
                       f"{latest}_kmeans_plus_scatter.png")
    if best_d:
        scatter_latest(best_d["labels"], f"{latest}: DBSCAN (eps={best_d['eps']}, min_samples={best_d['min_samples']}) sil={best_d['sil']:.3f}",
                       f"{latest}_dbscan_scatter.png")

    db = s_latest[s_latest["method"]=="dbscan"].dropna(subset=["eps","min_samples","sil"])
    if not db.empty:
        piv = db.pivot_table(index="min_samples", columns="eps", values="sil", aggfunc="max")
        plt.figure(figsize=(8,4.8))
        im = plt.imshow(piv.values, aspect="auto")
        plt.colorbar(im, label="Silhouette")
        plt.yticks(range(len(piv.index)), piv.index)
        plt.xticks(range(len(piv.columns)), piv.columns)
        plt.xlabel("eps"); plt.ylabel("min_samples")
        plt.title(f"DBSCAN silhouette (year {latest})")
        plt.tight_layout(); plt.savefig(os.path.join(FIG_DIR, f"{latest}_dbscan_heatmap.png"), bbox_inches="tight"); plt.show()

print("Figures saved to:", FIG_DIR)
```
